```python
maximumCv = [19.6257, 18.8007, 17.2829, 15.4828]
maximumX = [131.616, 93.62, 65.4225, 34.2343]
L = [50, 40, 30, 20]

from math import log
lnL = [log(y,np.e) for y in L]
lnCv = [log(y,np.e) for y in maximumCv]
lnX = [log(y,np.e) for y in maximumX]
```

```python
slopeC, interceptC, r_value, p_value, std_err = stats.linregress(lnCv, lnL)

print("The heat capacity line of best fit has parameters")
print("Slope: ", slopeC)
print("Intercept: ", interceptC)

lineC = []
for c in lnCv: 
    lineC.append(slopeC * c + interceptC)
    
```

$[C_{v} \sim L ^{-\gamma}]$      $\rightarrow$      $[\ln(C_{v}) \sim \beta \ln(L)]$

```python
plt.scatter(lnL, lnCv,label='ln(Cv) = 3.776 * ln(L) - 7.359')
plt.plot(lineC, lnCv)

plt.title('Heat Capactiy Scaling')
plt.xlabel('ln(L)')
plt.ylabel('ln(Cv)')
plt.legend()

plt.show()
```

The equation of the line of best fit is: $Cv = slope* ln(L) + intercept$

My results indicate that the slope is 0.216702, ie, the critical exponent is 0.2167.

```python
slopeX, interceptX, r_value, p_value, std_err = stats.linregress(lnX, lnL)

print("The heat capacity line of best fit has parameters")
print("Slope: ", slopeX)
print("Intercept: ", interceptX)

lineC = []
for c in lnX: 
    lineC.append(slopeX * c + interceptX)
```

$[X \sim L ^{\gamma}]$      $\rightarrow$      $[\ln(X) \sim \gamma \ln(L)]$

```python
plt.scatter(lnL, lnX)
plt.plot(lineC, lnX, label = 'ln(X) =  0.6862 * ln(L) - 0.5599')

plt.title('Magnetic Susceptability Scaling')
plt.xlabel('ln(L)')
plt.ylabel('ln(X[Tc])')
plt.legend()

plt.show()
```
